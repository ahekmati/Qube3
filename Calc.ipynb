```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return (x**3 + x) * sqrt(x)

# Generate x values, avoiding 0 and multiples of pi to prevent division by zero
x = np.linspace(-10, 10, 1000)
x = x[np.abs(np.sin(x)) > 0.01]  # Avoid division by values too close to zero

# Compute y values
y = f(x)

# Plot the function
plt.figure(figsize=(15, 6))
plt.plot(x, y, label=r'$f(x) = \frac{x^2}{\sin(x)}$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-100, 100)  # Limit y-range for better visualization
plt.title(r'Graph of $f(x) = \frac{x^2}{\sin(x)}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return (x**3 + x) * np.sqrt(x)

# Generate x values (only non-negative because sqrt(x) is undefined for x < 0 in reals)
x = np.linspace(0.01, 10, 1000)  # Start just above 0 to avoid sqrt(0) * 0

# Compute y values
y = f(x)

# Plot the function
plt.figure(figsize=(12, 6))
plt.plot(x, y, label=r'$f(x) = (x^3 + x)\sqrt{x}$', color='darkorange')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title(r'Graph of $f(x) = (x^3 + x)\sqrt{x}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**3 + x) * np.sqrt(x)  # original function

def F(x):
    return (2/9) * x**(9/2) + (2/5) * x**(5/2)  # antiderivative

x = np.linspace(0.01, 5, 500)

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='f(x) = (x³ + x)√x', color='blue')
plt.plot(x, F(x), label='F(x) = ∫f(x)dx', color='green', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparing f(x) and its Antiderivative')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values for plotting
x_exp = np.linspace(-2, 2, 400)       # For e^x, x can be any real number
x_log = np.linspace(0.01, 5, 400)     # For ln(x), x must be > 0

# Calculate y values
y_exp = np.exp(x_exp)
y_log = np.log(x_log)

# Create the plots
plt.figure(figsize=(14, 6))

# Plot e^x
plt.subplot(1, 2, 1)
plt.plot(x_exp, y_exp, label=r'$e^x$')
plt.title(r'Plot of $e^x$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Plot ln(x)
plt.subplot(1, 2, 2)
plt.plot(x_log, y_log, color='orange', label=r'$\ln(x)$')
plt.title(r'Plot of $\ln(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Demonstrate the inverse relationship
x_vals = [0.5, 1, 2, 3, 5]
print("x\t\te^x\t\tln(e^x)")
for x in x_vals:
    exp_val = np.exp(x)
    log_val = np.log(exp_val)
    print(f"{x:.2f}\t\t{exp_val:.4f}\t\t{log_val:.4f}")
```
