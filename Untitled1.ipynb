```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 2*x**4 + 8*x

# Generate x values
x = np.linspace(-3, 3, 400)  # You can adjust the range as needed

# Compute y values
y = f(x)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y, label=r'$y=2x^4+8x$')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.title("Graph of $y=2x^4+8x$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sympy import symbols, diff, solve, pprint

# Define the symbol and function
x = symbols('x')
y = 2*x**4 + 8*x

# Find the first derivative
y_prime = diff(y, x)
print("First derivative:")
pprint(y_prime)

# Find the second derivative
y_double_prime = diff(y_prime, x)
print("\nSecond derivative:")
pprint(y_double_prime)

# Solve y' = 0 for critical points
critical_points = solve(y_prime, x)
print("\nCritical points (solutions to y' = 0):")
print(critical_points)

# You can also print y values at the critical points
print("\nFunction values at critical points:")
for cp in critical_points:
    print(f"x = {cp}, y = {y.subs(x, cp)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.arctan(x**2)

# Generate x values
x = np.linspace(-4, 4, 400)

# Compute y values
y = f(x)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y, label=r'$y=\arctan(x^2)$')
plt.axhline(np.pi/2, color='red', linestyle='--', linewidth=1, label=r'$y=\frac{\pi}{2}$')
plt.title("Graph of $y=\\arctan(x^2)$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**2 - 1) / (x + 2)**2

# Avoid the vertical asymptote at x = -2 by not including it in x-values
x = np.linspace(-6, 2, 400)
x = x[x != -2]  # Remove x = -2 just to avoid division by zero

plt.figure(figsize=(8,5))
plt.plot(x, f(x), label=r'$y=\frac{x^2-1}{(x+2)^2}$')
plt.axhline(0, color='gray', linewidth=0.7)
plt.axvline(0, color='gray', linewidth=0.7)
plt.axvline(-2, color='red', linestyle='--', label='Vertical Asymptote x = -2')
plt.title('Graph of $y = \\frac{x^2-1}{(x+2)^2}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x + 2)**2 / (x**2 - 1)

x = np.linspace(-5, 5, 500)
x = x[np.abs(x - 1) > 0.02]   # avoid x=1
x = x[np.abs(x + 1) > 0.02]   # avoid x=-1

plt.figure(figsize=(8,6))
plt.plot(x, f(x), label=r'$y = \frac{(x+2)^2}{x^2 - 1}$')
plt.axvline(1, color='r', linestyle='--', label='Vertical Asymptote x = 1')
plt.axvline(-1, color='r', linestyle='--', label='Vertical Asymptote x = -1')
plt.axhline(0, color='k', linestyle=':')
plt.axhline(1, color='g', linestyle='--', label='Horizontal Asymptote y = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of $y = \\frac{(x+2)^2}{x^2 - 1}$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x + 2)**2 / (x**2 - 1)

def df(x):
    # Derivative computed manually:
    # f'(x) = (2(x+2)(x^2-1) - (x+2)^2*2x) / (x^2-1)^2
    numerator = 2*(x + 2)*(x**2 - 1) - 2*x*(x + 2)**2
    denominator = (x**2 - 1)**2
    return numerator / denominator

x = np.linspace(-5, 5, 1000)
x = x[(np.abs(x - 1) > 0.02) & (np.abs(x + 1) > 0.02)] # remove vertical asymptotes

y = f(x)
dy = df(x)

increasing = dy > 0
decreasing = dy < 0

plt.figure(figsize=(10,6))
plt.plot(x[increasing], y[increasing], color='green', label='Increasing')
plt.plot(x[decreasing], y[decreasing], color='red', label='Decreasing')
plt.axvline(1, color='gray', linestyle='--', label='x = 1 asymptote')
plt.axvline(-1, color='gray', linestyle='--', label='x = -1 asymptote')
plt.axhline(1, color='purple', linestyle='--', label='y = 1 asymptote')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Intervals of Increase/Decrease for $f(x) = \\frac{(x+2)^2}{x^2-1}$')
plt.legend()
plt.grid(True)
plt.show()
```
