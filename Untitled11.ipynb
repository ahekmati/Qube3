```python
import numpy as np
import matplotlib.pyplot as plt

# Define a 2D orthogonal matrix (rotation by 45 deg)
theta = np.radians(45)
Q = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# Original basis vectors (columns of the 2x2 identity)
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Transform basis vectors using the orthogonal matrix
Te1 = Q @ e1
Te2 = Q @ e2

# Visualize
plt.figure(figsize=(6, 6))
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='red', label='Original e1')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original e2')
plt.quiver(0, 0, Te1[0], Te1[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Transformed e1')
plt.quiver(0, 0, Te2[0], Te2[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='Transformed e2')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend()
plt.gca().set_aspect('equal')
plt.title("Visualization of a 2D Orthogonal Matrix (Rotation by 45Â°)")
plt.show()
```
