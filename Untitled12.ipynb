```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# Download daily close prices for QQQ, AAPL, MSFT, GOOG, SPY (10 years history)
tickers = ["QQQ", "AAPL", "MSFT", "GOOG", "SPY"]
data = yf.download(tickers, start="2024-01-01", end="2025-11-01")["Close"]

# Remove rows with missing values (for PCA)
prices = data.dropna()

# Calculate daily returns for PCA universe
returns = prices[["AAPL", "MSFT", "GOOG"]].pct_change().dropna()

# --- PCA rolling signal setup ---
window = 20  # lookback window for PCA and rolling mean
signals = []
pca_series = []

for i in range(window, len(returns)):
    pca = PCA(n_components=1)
    rolling_returns = returns.iloc[i-window:i]
    pca.fit(rolling_returns)
    PC1_scores = rolling_returns @ pca.components_[0]  # Project returns onto PC1 axis
    curr_PC1 = PC1_scores.iloc[-1]
    ma_PC1 = PC1_scores.rolling(window=5).mean().iloc[-2]  # lagged moving average

    signal = 1 if curr_PC1 > ma_PC1 else -1 if curr_PC1 < ma_PC1 else 0
    signals.append(signal)
    pca_series.append(curr_PC1)

signal_index = returns.index[window:]
signals = pd.Series(signals, index=signal_index)

# --- Return alignments for backtest ---
qqq_rets = prices["QQQ"].pct_change().reindex(signal_index).fillna(0)
spy_rets = prices["SPY"].pct_change().reindex(signal_index).fillna(0)

# Strategy positions and PnL
position = signals.shift(1).fillna(0)  # trade signal for next day
pnl = position * qqq_rets
cum_pnl = pnl.cumsum()

spread_position = position
spread_pnl = spread_position * (qqq_rets - spy_rets)
cum_spread_pnl = spread_pnl.cumsum()

# Print all trade signal dates and QQQ prices
trades = signals.diff().fillna(0).astype(bool)
trade_dates = signal_index[trades]
trade_prices = prices["QQQ"].reindex(trade_dates)

for date, sig, price in zip(trade_dates, signals.loc[trade_dates], trade_prices):
    action = "LONG" if sig == 1 else "SHORT" if sig == -1 else "EXIT"
    print(f"{date.date()} | {action} QQQ @ ${price:.2f}")

print("\nFinal QQQ strat PnL (1 share):", cum_pnl.iloc[-1])
print("Final QQQ-SPY spread strat PnL (1:1 share):", cum_spread_pnl.iloc[-1])
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# Set parameters
start_date = "2020-01-01"
end_date = "2025-01-01"
investment = 10000
stocks = ["AAPL", "MSFT", "GOOG"]

# Download data
data = yf.download(stocks, start=start_date, end=end_date)["Close"].dropna()

# Calculate returns for PCA
returns = data.pct_change().dropna()

# Fit PCA for entire period and get PC1 weights
pca = PCA(n_components=1)
pca.fit(returns)
pc1_weights_raw = pca.components_[0]
pc1_weights = pc1_weights_raw / np.sum(np.abs(pc1_weights_raw))

# Equal weight for 3 stocks
equal_weights = np.full(len(stocks), 1/len(stocks))

# Initial allocation (number of shares)
init_prices = data.iloc[0]
shares_equal = (investment * equal_weights) / init_prices.values
shares_pc1 = (investment * pc1_weights) / init_prices.values

# Daily portfolio values
values_equal = (data * shares_equal).sum(axis=1)
values_pc1 = (data * shares_pc1).sum(axis=1)

# Total returns
ret_equal = (values_equal.iloc[-1] - values_equal.iloc[0]) / values_equal.iloc[0]
ret_pc1 = (values_pc1.iloc[-1] - values_pc1.iloc[0]) / values_pc1.iloc[0]

# Annualized volatility
vol_equal = values_equal.pct_change().std() * np.sqrt(252)
vol_pc1 = values_pc1.pct_change().std() * np.sqrt(252)

print("PC1 Weights (Normalized):")
for s, w in zip(stocks, pc1_weights):
    print(f"  {s}: {w:.3f}")

print("\nPortfolio comparison")
print(f"Equal Weight: Total Return={ret_equal:.2%}, Annualized Volatility={vol_equal:.2%}")
print(f"PC1 Weighted: Total Return={ret_pc1:.2%}, Annualized Volatility={vol_pc1:.2%}")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras

# 1. Download basket prices and target ETF
basket = ["AAPL", "MSFT", "GOOG"]
target = "TQQQ"
start, end = "2020-01-01", "2025-01-01"
data = yf.download(basket + [target], start=start, end=end)["Close"].dropna()

# 2. Create returns and fit PCA
returns = data[basket].pct_change().dropna()
pca = PCA(n_components=1)
pca.fit(returns)
pc1_signal = returns @ pca.components_[0]

# 3. Prepare supervised data for LSTM (use next day's TQQQ return as the label)
# Add PC1 as only input feature
df = pd.DataFrame({
    "PC1": pc1_signal,
    "TQQQ": data[target].pct_change().reindex(pc1_signal.index)
}).dropna()

# Create binary buy signal: buy if next day's TQQQ return > 0
df["BuySignal"] = (df["TQQQ"].shift(-1) > 0).astype(int)

# 4. Split to train/test sets and scale
split = int(len(df) * 0.8)
X = df[["PC1"]].values
y = df["BuySignal"].values
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test = X_scaled[:split], X_scaled[split:]
y_train, y_test = y[:split], y[split:]

# 5. Reshape for LSTM [samples, time steps, features]
X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# 6. Train LSTM: 1 input feature, 1 output, binary classification
model = keras.Sequential([
    keras.layers.LSTM(32, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train_lstm, y_train, epochs=8, batch_size=16, verbose=0)

# 7. Predict signals using LSTM vs. plain PCA
lstm_probs = model.predict(X_test_lstm).flatten()
lstm_signals = (lstm_probs > 0.5).astype(int)
pca_signals = (X_test.flatten() > 0).astype(int)

# 8. Compare signal hit rates (accuracy) on test set
from sklearn.metrics import accuracy_score

print("Test accuracy (buy signals):")
print(f"PCA-only: {accuracy_score(y_test, pca_signals):.2%}")
print(f"LSTM (trained on PC1 signal): {accuracy_score(y_test, lstm_signals):.2%}")

# Optionally, print the signal series to visually compare
import matplotlib.pyplot as plt

plt.plot(y_test, label='Actual')
plt.plot(pca_signals, label='PCA signals')
plt.plot(lstm_signals, label='LSTM signals')
plt.legend()
plt.title('Buy Signal Comparison (PCA vs LSTM)')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
from ta.momentum import RSIIndicator
from ta.volatility import BollingerBands

basket = ["AAPL", "MSFT", "GOOG"]
target = "TQQQ"
start, end = "2020-01-01", "2025-01-01"
data = yf.download(basket + [target], start=start, end=end)["Close"].dropna()

returns = data[basket].pct_change().dropna()
pca = PCA(n_components=1)
pca.fit(returns)
pc1_signal = returns @ pca.components_[0]

# Feature engineering for entire target ETF
features = pd.DataFrame(index=returns.index)
features["PC1"] = pc1_signal
features["PC1_Lag1"] = pc1_signal.shift(1)
features["PC1_Lag2"] = pc1_signal.shift(2)
features["TQQQ_Return"] = data[target].pct_change().reindex(features.index)
features["TQQQ_MA5"] = data[target].rolling(5).mean().reindex(features.index)
features["TQQQ_MA20"] = data[target].rolling(20).mean().reindex(features.index)
features["TQQQ_RSI"] = RSIIndicator(data[target].reindex(features.index)).rsi()
bb = BollingerBands(data[target].reindex(features.index))
features["TQQQ_BB_Width"] = bb.bollinger_hband() - bb.bollinger_lband()  # Volatility

# Drop resulting NaNs
features = features.dropna()
labels = (features["TQQQ_Return"].shift(-1) > 0).astype(int)  # Next-day "buy" label

# Prepare rolling window sequences for LSTM (e.g., 10 days)
N = 10  # Sequence length
def create_sequences(X, y, N):
    Xs, ys = [], []
    for i in range(len(X) - N):
        Xs.append(X[i:(i+N)])
        ys.append(y[i+N])
    return np.array(Xs), np.array(ys)

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(features.values)
X_seq, y_seq = create_sequences(X_scaled, labels.values, N)

# Train/test split
split = int(len(X_seq) * 0.8)
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]

# LSTM model
model = keras.Sequential([
    keras.layers.LSTM(32, input_shape=(X_train.shape[1], X_train.shape[2])),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20, batch_size=16, verbose=0)

lstm_probs = model.predict(X_test).flatten()
lstm_signals = (lstm_probs > 0.5).astype(int)

# PCA-only signals: simple threshold on latest PC1
pc1_test = X_test[:, -1, 0]
pca_signals = (pc1_test > 0).astype(int)

# Get test dates and prices
test_dates = features.index[split+N:]
test_prices = data[target].reindex(test_dates)

print("\nLast 10 trade signals and prices:")

print("\nPCA-only signals (date, signal, price):")
for date, sig, price in zip(test_dates[-10:], pca_signals[-10:], test_prices[-10:]):
    if sig:
        print(f"{date.date()}: Buy at ${price:.2f}")

print("\nLSTM signals (date, signal, price):")
for date, sig, price in zip(test_dates[-10:], lstm_signals[-10:], test_prices[-10:]):
    if sig:
        print(f"{date.date()}: Buy at ${price:.2f}")

from sklearn.metrics import accuracy_score
print(f"\nPCA Accuracy: {accuracy_score(y_test, pca_signals):.2%}")
print(f"LSTM Accuracy: {accuracy_score(y_test, lstm_signals):.2%}")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# -----------------------------
# Parameters
# -----------------------------
ticker = "QQQ"
period = "6mo"
interval = "1d"
fast_ema = 7
slow_ema = 12
sma_period = 40

# -----------------------------
# Download Data
# -----------------------------
df = yf.download(ticker, period=period, interval=interval)
print("[*] Downloaded Data")
print("Columns in DataFrame:", df.columns)

# Flatten MultiIndex if needed
if isinstance(df.columns, pd.MultiIndex):
    df.columns = ['_'.join(col).strip() for col in df.columns.values]

# Ensure column names are lowercase
df.columns = [c.lower() for c in df.columns]

# Check for 'close' column
if 'close' not in df.columns:
    raise KeyError("No 'close' column found in the data.")

# -----------------------------
# Compute returns
# -----------------------------
df['returns'] = df['close'].pct_change()

# Drop NA for PCA
returns_dropna = df[['returns']].dropna()

# -----------------------------
# Standardize returns
# -----------------------------
scaler = StandardScaler()
returns_std = scaler.fit_transform(returns_dropna)

# -----------------------------
# PCA (example, only 1 feature here)
# -----------------------------
pca = PCA(n_components=1)
pca.fit(returns_std)
print("PCA Explained Variance Ratio:", pca.explained_variance_ratio_)

# -----------------------------
# Moving averages
# -----------------------------
df['ema7'] = df['close'].ewm(span=fast_ema, adjust=False).mean()
df['ema12'] = df['close'].ewm(span=slow_ema, adjust=False).mean()
df['sma40'] = df['close'].rolling(sma_period).mean()

# -----------------------------
# Generate signals
# -----------------------------
df['signal'] = 0

for i in range(1, len(df)):
    # 12 EMA < 40 SMA
    condition1 = df.loc[df.index[i], 'ema12'] < df.loc[df.index[i], 'sma40']
    # 7 EMA crosses above 12 EMA
    condition2 = (df.loc[df.index[i-1], 'ema7'] <= df.loc[df.index[i-1], 'ema12']) and \
                 (df.loc[df.index[i], 'ema7'] > df.loc[df.index[i], 'ema12'])
    
    if condition1 and condition2:
        df.loc[df.index[i], 'signal'] = 1  # Buy
        print(f"{df.index[i]} BUY @ Close={df.loc[df.index[i], 'close']}")
    # Optional: add sell condition
    elif df.loc[df.index[i], 'ema7'] < df.loc[df.index[i], 'ema12']:
        df.loc[df.index[i], 'signal'] = -1  # Sell
        print(f"{df.index[i]} SELL @ Close={df.loc[df.index[i], 'close']}")

print("Done.")
```
