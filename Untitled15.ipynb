```python
import numpy as np
import matplotlib.pyplot as plt

# Define a test function of two variables
def f(x, y):
    return x**2 - y**2

# Compute gradients
def df_dx(x, y):
    return 2*x

def df_dy(x, y):
    return -2*y

# Grid over which to plot
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Gradients
U = df_dx(X, Y)
V = df_dy(X, Y)

plt.figure(figsize=(7,6))
plt.contourf(X, Y, Z, cmap='coolwarm', levels=20, alpha=0.6) # Scalar field contours
plt.quiver(X, Y, U, V, color='black') # Gradient vectors
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scalar field & gradient vectors')
plt.show()
```
