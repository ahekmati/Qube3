```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (start from slightly above zero to avoid ln(0))
x = np.linspace(1.01, 50, 500)
y = x / np.log(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('x / ln(x)')
plt.title('Graph of y = x / ln(x)')
plt.grid(True)
plt.ylim(0, 30)  # Adjust as needed for preferred visualization
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# To avoid ln(0), start just above zero
x = np.linspace(0.01, 50, 500)
y = x * np.log(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('x * ln(x)')
plt.title('Graph of y = x * ln(x)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x range from -2*pi to 2*pi
x = np.linspace(-2*np.pi, 2*np.pi, 400)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.scatter(0, np.sin(0), color='red', label='sin(0)')
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.title('Graph of sin(x) with sin(0) Highlighted')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values (range from -5 to 2)
x = np.linspace(-5, 2, 400)
# Define y = (x + 3)e^x
y = (x + 3) * np.exp(x)

# Plot the graph
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='y = (x + 3) * e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = (x + 3) * e^x')
plt.axhline(0, color='gray', linestyle='--', alpha=0.7)
plt.axvline(0, color='gray', linestyle='--', alpha=0.7)
plt.grid(True)
plt.legend()
plt.show()
```
