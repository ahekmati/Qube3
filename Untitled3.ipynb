```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate e^-10
y = np.exp(-10)

# Create an array of x values (just for the plot)
x = np.linspace(-10, 10, 200)
y_vals = np.full_like(x, y)

plt.figure(figsize=(8, 2))
plt.plot(x, y_vals, label='y = e^{-10}')
plt.title('Graph of y = e^{-10} (constant function)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-2, 5, 400)
# Compute the function values
y = 3 * x**2 - x**3 - 1

# Plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="f(x) = 3x^2 - x^3 - 1")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.7)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = 3x^2 - x^3 - 1')
plt.legend()
plt.grid(True)
plt.show()
```
