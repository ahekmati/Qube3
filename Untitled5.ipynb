```python
# Simple Python script to visualize vector addition (u, v, and u+v)

import matplotlib.pyplot as plt
import numpy as np

# Define vectors
u = np.array([2, 3])
v = np.array([-1, 4])

# Create the resulting vector (u + v)
uv_sum = u + v

# Set up the plot
plt.figure(figsize=(6,6))
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)

# Plot the vectors from the origin
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u = (2,3)')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='v = (-1,4)')
plt.quiver(0, 0, uv_sum[0], uv_sum[1], angles='xy', scale_units='xy', scale=1, color='green', label='u + v = (1,7)')

# Configure the grid and labels
plt.xlim(-3, 4)
plt.ylim(-1, 8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title('Vector Visualization: u, v, and u + v')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Display the plot
plt.show()
```
