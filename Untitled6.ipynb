```python
import numpy as np
import matplotlib.pyplot as plt

# Generate grid points
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])

# 2x2 transformation matrix: rotation (45°) + shear
theta = np.pi / 4
shear = 0.5
A = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta) + shear, np.cos(theta)]])

# Apply transformation
transformed_points = A @ points

# Plot both sets
plt.figure(figsize=(7,7))
plt.scatter(points[0], points[1], color='blue', alpha=0.6, label='Original Grid')
plt.scatter(transformed_points[0], transformed_points[1], color='red', alpha=0.6, label='Transformed Grid')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.legend()
plt.title('Linear Algebra Transformation: Rotation + Shear')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# --- Parameters ---
# Define a 2x2 transformation matrix
A = np.array([[1, 0.5],
              [0.2, 1]])

# Create a grid of points
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

# Flatten the grid for transformation
points = np.vstack([X.flatten(), Y.flatten()])

# Define a few sample vectors to show transformation
vectors = np.array([[1, 0],
                    [0, 1],
                    [1, 1],
                    [-1, 1]]).T

# --- Animation setup ---
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_aspect('equal', 'box')
ax.set_title("Linear Transformation Animation")

# Initial scatter plot
grid_plot = ax.scatter(points[0], points[1], color='lightblue', s=10)
quivers = ax.quiver([0]*4, [0]*4, vectors[0], vectors[1], 
                    angles='xy', scale_units='xy', scale=1, color='red')

# --- Update function ---
def update(frame):
    # Interpolate transformation from identity to A
    t = frame / 100
    M = (1 - t) * np.eye(2) + t * A

    # Apply transformation
    transformed_points = M @ points
    transformed_vectors = M @ vectors

    # Update grid and vectors
    grid_plot.set_offsets(transformed_points.T)
    quivers.set_UVC(transformed_vectors[0], transformed_vectors[1])
    ax.set_title(f"Linear Transformation (t = {t:.2f})")
    return grid_plot, quivers

# --- Animate ---
ani = FuncAnimation(fig, update, frames=101, interval=50, blit=False)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values close to zero up to 10
x = np.linspace(0.001, 10, 500)
y = np.log(x)

plt.figure(figsize=(7, 5))
plt.plot(x, y, color='blue', label=r'$y = \ln(x)$')
plt.axvline(x=0, color='red', linestyle='--', label='x = 0 (asymptote)')
plt.title(r'Graph of $y = \ln(x)$ as $x \to 0^+$')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 3)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values, skip x=0 to avoid division by zero
x = np.linspace(-5, 5, 500)
x = x[x != 0]  # remove zero to prevent division error
y = x / x**3   # or y = 1 / x**2

plt.plot(x, y, label='y = x/x³')
plt.axhline(0, color='black', lw=1, ls='--')
plt.axvline(0, color='black', lw=1, ls='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y=x/x³')
plt.ylim(-2, 2)  # Set y-limit for better visualization
plt.legend()
plt.grid(True)
plt.show()
```
